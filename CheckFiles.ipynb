```python
import os
import glob

# 檢查 file 1-4 中有多少 xlsx 檔案
base_path = r"E:\EarthScienceFair_Data"
target_folders = ["1", "2", "3", "4"]

print("="*80)
print("檔案結構檢查")
print("="*80)

total_count = 0

for folder in target_folders:
    folder_path = os.path.join(base_path, folder)
    
    if not os.path.exists(folder_path):
        print(f"\n資料夾 {folder} 不存在！")
        continue
    
    print(f"\n資料夾 {folder}:")
    
    # 列出所有子資料夾
    subdirs = [d for d in os.listdir(folder_path) if os.path.isdir(os.path.join(folder_path, d))]
    print(f"  子資料夾: {subdirs}")
    
    # 找出所有 xlsx 檔案
    xlsx_files = glob.glob(os.path.join(folder_path, "**", "*.xlsx"), recursive=True)
    
    print(f"  xlsx 檔案數量: {len(xlsx_files)}")
    
    for f in xlsx_files:
        print(f"    - {os.path.relpath(f, base_path)}")
        total_count += 1

print("\n" + "="*80)
print(f"總計: {total_count} 個 xlsx 檔案")
print("="*80)

# 額外檢查：是否有其他副檔名的檔案
print("\n其他檔案類型:")
for folder in target_folders:
    folder_path = os.path.join(base_path, folder)
    if os.path.exists(folder_path):
        all_files = glob.glob(os.path.join(folder_path, "**", "*.*"), recursive=True)
        extensions = set([os.path.splitext(f)[1] for f in all_files])
        print(f"  資料夾 {folder}: {sorted(extensions)}")
```
